# 2. Análise Qualitativa
"Dados qualitativos são informações que descrevem **qualidades ou categorias** em vez de valores numéricos. Eles são usados para classificar ou rotular elementos em um conjunto de dados"

## 2.1 Contagem de tipos de Pokémon
Vamos contar quantos Pokémons únicos existem na base de dados, e quais são os diferentes tipos de Pokémon, e quantos existem de cada tipo.

```python
df['Name'].nunique() # A função Nunique retorna o total de elementos unicos encontrados na coluna
```

```python
df['Type'].value_counts() # Essa função retorna o total de ocorrencias em uma coluna do dataframe
```

## 2.2 Habilidades Únicas
Agora vamos repetir o que fizemos para os tipos de Pokémon, utilizando a função *value_counts()* para contar o total de habilidades únicas e também verificar a habilidade mais comum entre os Pokémon utilizando *idmax()* na series retornada, ou seja, o valor que mais se repete nessa coluna.

```python
df['Abilities'].nunique() # Aqui, repete-se o processo de verificar o total de valores unicos em uma coluna
```

```python
df['Abilities'].value_counts().idxmax() # Retorna o nome do elemento que mais se repete na coluna
```

## 2.3 Taxa de captura
Aqui, aplicaremos outras funções para verificar a taxa de captura mais comum, e qual o Pokémon com a maior taxa de captura, e qual tem a menor taxa de captura, utilizando filtragem, é possível notar que há mais de um Pokémon que se encaixa nesses critérios

```python
df['Catch rate'].mode() # Por se tratar de numeros inteiros, utilizamos apenas a moda matematica (numero mais frequente) para descobrir qual a taxa de captura mais comum
```

### Filtrando Pokémons baseado na taxa de captura

```python
greatest_cr = df['Catch rate'] == df['Catch rate'].max() # Filtro de maior taxa de captura
lowest_cr = df['Catch rate'] == df['Catch rate'].min()
```

```python
# Aqui sera impressa a lista apenas de Pokemons que tem a menor taxa de captura
df[lowest_cr]
```

```python
# Aqui sera impressa a lista de Pokemons com a maior taxa de captura
df[greatest_cr]
```

# 3. Análise Quantitativa
"Quando analisamos uma variável qualitativa, basicamente construímos uma **distribuição de frequência**. No entanto, ao explorarmos variáveis quantitativas, temos condições de empregar algumas medidas descritivas que sintetizam as características da distribuição"

## 3.1 Características Físicas dos Pokémons
Para ter uma análise quantitativa mais dinâmica, vamos separar dados entre características físicas e estatísticas de batalha. Primeiro vamos analisar a altura e o peso dos Pokémons, e analisar se há alguma relação entre altura e peso

```python
weight_mean = df['Weight'].mean() # Calculando a media de peso ddf['Body Type'] = os Pokemons
height_mean = df['Height'].mean()
print('O peso médio é:', weight_mean)
print('A altura média é:', height_mean)
```

### Relacionando Altura e Peso
É possível criar uma correlação entre peso utilizando a fórmula do IMC (Peso/Altura^2)


```python
df['IMC'] = df['Weight']/(df['Height'] ** 2)
df.head(5)
```

## 3.2 Estatísticas de Batalha
Agora, as estatísticas de batalha serão analisadas, calcularemos o maior valor total de estatísticas (*soma de hp, ataque, defesa, etc.*), Os Pokémons mais rápidos e a variação de HP entre os diferentes tipos de Pokémon

```python
# Criando uma coluna que soma todas as estatisticas de batalha de um Pokémon
df['Overall Power'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Def'] + df['Sp. Atk'] + df['Speed']

# Criando um filtro para descobrir os que se encaixam nos criterios dados
greatest_power = df['Overall Power'] == df['Overall Power'].max()
df[greatest_power]
```

```python
# Organizando os valores pela velocidade de forma descendente e mostrando apenas os 5 primeiros elementos do dataframe
df.sort_values('Speed', ascending = False).head(5)
```

### Análise da média de HP
Utilizando o método *groupby()*, vamos criar verificar a média de HP para cada tipo de Pokémon

```python
df.groupby('Type')['HP'].mean()
```

# 4. Demonstrando Estatísticas
Os dados serão filtrados, e gráficos serão criados para demonstrar algumas estatísticas sobre os Pokémon

##  4.1 Filtragem de dados
Vamos criar filtros para os Pokémon do tipo "Fire" e para os Pokémon com ataque acima de 100

```python
fire_type = df['Type'] == 'Fire'
atk_greater_100 = df['Attack'] > 100

# Aplica ambos os filtros e mostra apenas os Pokémon do tipo fogo com ataque acima de 100
df[fire_type & atk_greater_100]
```

## 4.2 Criação de Gráficos
Serão feitos 2 gráficos, um gráfico de barras para a quantidade de Pokémon por tipo e um gráfico de dispersão monstrando a relação de Ataque e Defesa

### Gráfico de Tipo de Pokémon

```python
type_counts = df['Type'].value_counts()
colors = ['blue','red','green']

plt.figure(figsize=(25, 6)) # Define o tamanho da figura (imagem do grafico)
type_counts.plot(kind='bar') # Cria o grafico de barra
plt.title('Quantidade de Pokémon por Tipo') # Titulo
plt.xlabel('Tipo de Pokémon') # Rotulo do eixo x
plt.ylabel('Quantidade') # Rotulo do eixo y
plt.xticks(rotation=90,ha='center',fontsize=7) # Deixa os nomes dos tipos mais legiveis

plt.show()
```

### Gráfico de Dispersão de Pokémon

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Attack'], df['Defense'], alpha=0.5, c='blue') # Cria um grafico de dispersao que recebe os valores das colunas attack e defense como parametros
plt.title('Relação Ataque / Defesa')
plt.xlabel('Ataque')
plt.ylabel('Defesa')
plt.grid(True)
plt.show()
```
